# Importing Libraries

```python
import pandas as pd
import numpy as np
import os.path as osp

pd.options.display.max_columns = 100
pd.options.display.float_format = "{:,.2f}".format
```

# Definitions

```python
DATA_PATH = osp.join('..', 'data','data_cleaned.csv')
TARGET = 'qt_adjuvante_-_pratica_clinica'

categorical_features=[
      'etilismo'
    , 'tabagismo'
    , 'status_menopausa'
    , 'lateralidade'
    , 'histologia'
    , 'multifocal'
    , 'estadiamento_patologico'
    , 'estadio_n'
    , 'ihq_lfn'
    , 'ec'
    , 'gh'
    , 'risco_clinico'
    , 'ial'
    , 'cirurgia'
    , 'bls'
    , 'disseccao_axilar'
    , 'regime_qt'
    , 'te_adjuvante'
    , 'switch'
    , 'te'
    , 'supressao_ovariana'
    , 'monitoracao_de_estradiol_ultrass'
    , 'supressao_ovariana_adequada'
    , 'radioterapia_adjuvante'
    , 'recorrencia'
    , 'sitio'
    # , 'obito'
    ]

numeric_features = [
      'altura'
    , 'idade_ao_diagnostico'
    , 'imc'
    , 'ki67'
    , 'odxrs'
    , 'peso'
    , 're'
    , 'rp'
    , 'tamanho'
    , 'tils'
    , 'tempo_para_cirurgia'
 ]

selected_features = set(categorical_features + numeric_features + [TARGET])
```

# Reading Data

```python
df = pd.read_excel('../data/Planilha Dados clínicos.xlsx', sheet_name = 0, header = 1)
df.head()
```

# Renaming columns

```python
# !pip install unidecode
import unidecode

def standardize_column_name(col):
    col = unidecode.unidecode(col)
    return col.replace(' ', '_').lower()

df.columns = [standardize_column_name(col) for col in df.columns]
```

# Data Cleaning

## Standardize Null, True and False

```python
standardize_values_dict = {
    'etilismo': { 
          0: 999 #Desconhecido
        , 1: 0 #Não 
        , 2: 1 #Sim
    }
    , 'tabagismo': { 
          0: 999 #Desconhecido
        , 1: 0 #Não 
        , 2: 1 #Sim
    }
    , 'status_menopausa': {
        3:999 #Desconhecido
    }
    , 'lateralidade': {
        4:999 #Desconhecido
    }
    , 'ihq_lfn': {
          0:1 #. Sim
        , 1:0 #. Não
    }
    , 'bls': {
          0: 1 #. Sim
        , 1: 0 #. Não
    }
    , 'te_adjuvante':{
        3:0 #Não
    }
    , 'supressao_ovariana_adequada': {
          0: 1 #. Sim
        , 1: 0 #. Não
    }
}
```

```python
for col, values in standardize_values_dict.items():
    df[col] = df[col].replace(values)
```

## Treating missing rows

```python
# Drop rows with all missing values
df.dropna(how = 'all', inplace = True)

# Fill missing values with 999
df = df.fillna(999)
```

# Feature Engineering

```python
# replacing 999 with NaN
df[['peso','altura']] = df[['peso','altura']].replace(999, np.nan)

# Recalculate IMC
df['imc'] = df['peso'] / ((df['altura']/100) ** 2)

# Calculate time until surgery
df[['data_diagnostico', 'data_da_cirurgia']] = df[['data_diagnostico', 'data_da_cirurgia']].apply(pd.to_datetime, errors = 'coerce')
df['tempo_para_cirurgia'] = (df['data_da_cirurgia'] - df['data_diagnostico']).dt.days
```

# Adjusting Dtypes

```python
# Fill missing values with 999
df = df.fillna(999)

dtypes = {
    'prontuario': int
    , 'idade_ao_diagnostico': int
    , 'peso': float
    , 'altura': float
    , 'imc': float # will be recalculated
    , 'etilismo': int
    , 'tabagismo': int
    , 'status_menopausa': int
    , 'lateralidade': int
    , 'histologia': int
    , 'multifocal': int
    , 'estadiamento_patologico': int
    , 'estadio_n': int
    , 'ihq_lfn': int
    , 'tamanho': float
    , 'ec': int
    , 'gh': int
    , 'risco_clinico': int
    , 're': float
    , 'rp': float
    , 'ki67': float
    , 'tils': float
    , 'ial': int
    , 'cirurgia': int
    , 'bls': int
    , 'disseccao_axilar': int
    , 'odxrs': int
    , 'qt_adjuvante_-_pratica_clinica': int
    , 'te_adjuvante': int
    , 'supressao_ovariana_adequada': int
    , 'regime_qt' : int
    , 'switch' : int
    , 'te' : int
    , 'supressao_ovariana' : int
    , 'monitoracao_de_estradiol_ultrass' : int
    , 'radioterapia_adjuvante' : int
    , 'recorrencia' : int
    , 'sitio' : int
}

# Apply dtypes
df = df.astype(dtypes)

```

## Preprocessing

```python
# dropping missing values on the target
df.dropna(subset=[TARGET], inplace=True)
```

```python
# defining categorical features
df[categorical_features] = df[categorical_features].fillna('missing')
df[categorical_features] = df[categorical_features].astype(str)

# defining numeric features
df[numeric_features] = df[numeric_features].astype('float')

# defining o target
df[TARGET] = df[TARGET].astype('int')
```

```python
# replacing 999 with NaN
df.replace([999,'999','999.0'], np.nan, inplace = True)
```
